# Sorting things Chronologically

I'm trying to have the raceId, driverId be based more on chronology instead of starting with the 2009 season and going back to fill things out.

```python
import numpy as np
import pandas as pd
```

```python
drivers = pd.read_csv("f1db/chrono/drivers.csv")
```

```python
drivers.head()
```

```python
driverId2 = drivers[["driverId", "chronoDriverId"]]
driverId2.head()
```

```python
races = pd.read_csv("f1db/chrono/races.csv")
```

```python
races.head()
```

```python
raceId2 = races[["raceId","chronoRaceId"]]
raceId2.head()
```

```python
const_results = pd.read_csv("f1db/chrono/constructor_results.csv")
```

```python
const_results.head()
```

```python
k = const_results.join(raceId2.set_index("raceId"), on="raceId")
```

```python
k
```

```python
k.to_csv("f1db/chrono/constructor_results_2.csv")
```

Let's see now which ones we have to change ... 

* `driver_standings`, `results`, `qualifying`, `pit_stops`, `lap_times` needs both raceId and driverId
* `constructor_standings` needs raceId

```python
const_standings = pd.read_csv("f1db/chrono/constructor_standings.csv")
const_standings.head(20)
```

```python
a = const_standings.join(raceId2.set_index("raceId"), on="raceId")
```

```python
a.head(20)
```

```python
a.tail(20)
```

```python
a.to_csv("f1db/chrono/constructor_standings_2.csv")
```

```python
df = pd.read_csv("f1db/chrono/driver_standings.csv")
df2 = df.join(raceId2.set_index("raceId"), on="raceId").join(driverId2.set_index("driverId"), on ="driverId")
```

```python
df2.tail(25)
```

```python
df2.to_csv("f1db/chrono/driver_standings_2.csv")
```

```python
filename = "results"
df = pd.read_csv("f1db/chrono/"+filename+".csv")
df2 = df.join(raceId2.set_index("raceId"), on="raceId").join(driverId2.set_index("driverId"), on ="driverId")
df2.to_csv("f1db/chrono/"+filename+"_2.csv")
```

```python
files = ["qualifying", "pit_stops", "lap_times"]

for i in files:
    print(i)
    filename = i 
    df = pd.read_csv("f1db/chrono/"+filename+".csv")
    df2 = df.join(raceId2.set_index("raceId"), on="raceId").join(driverId2.set_index("driverId"), on ="driverId")
    df2.to_csv("f1db/chrono/"+filename+"_2.csv")
```
