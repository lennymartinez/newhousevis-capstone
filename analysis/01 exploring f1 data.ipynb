```python
def getPlace(driverId, position):
    results = pd.read_csv("f1db/results.csv")
    driverGrouped = results.groupby("driverId")
    driver = driverGrouped.get_group(driverId)
    driverStandings = driver.positionOrder.value_counts()
    return driverStandings[position]
```
